```python
print('Welcome to the Quiz Tournament! The game where you need to get as many questions right as possible!.\n')

# variables
again = True # this will be used for testing if the user wants to play the game again

# check if they want to play
playing = input('Would you like to enter the tournament?\nEnter "yes" or "no": ')

# validate their response; checking for invalid input
while (playing != 'yes' and playing != 'no'):
    playing = input('/nInvalid!/nEnter "yes" or "no": ')

# quit game if they choose 'no'
if playing != 'yes':
    quit()

print('\nGreat! We have a list of topics to choose from. We have a chemistry quiz and a history quiz.')

# topic choice
topic = input('\nWhich would you like to pick?\nEnter "chemistry" or "history": ')

# validate their response; checking for invalid input
while (topic != 'chemistry' and topic != 'history'):
    topic = input('\nInvalid!\nEnter "chemistry" or "history": ')

print('\nWonderful! Good luck out there, make sure to do your best!')

# this will make sure to use chemistry questions for the chemistry quiz
while topic == 'chemistry' and again == True:
    # gives user prompt to answer
    answer = input('\nProblem 1:\nElements consist of a single type of atom.'
                   '\nEnter "t" for true or "f" for false: ')
    # validate their response; checking for invalid input
    while (answer != 't' and answer != 'f'):
        answer = input('\nInvalid!\nEnter "t" or "f": ')
    
    # printing out responses based on answer given
    if answer == 't':
        print('\nGreat job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')

    answer2 = input('\nProblem 2: As you move left along with the periodic table, the atomic' 
                ' number of the elements tends to: '
                '\na: increase'
                '\nb: decrease'
                '\nc: stay the same'
                '\nAnswer: ')

    while (answer2 != 'a' and answer2 != 'b' and answer2 != 'c'):
        answer2 = input('\nInvalid!\nEnter "a", "b", or "c" : ')

    # this will be used to print a different response based on if they get multiple questions right consecutively
    if (answer == 't' and answer2 == 'b'):
        print('\nThat\'s two in a row!')
    elif answer2 == 'b':
        print('Great job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')

    answer3 = input('\nProblem 3: Which of the following is NOT an indicator of a Chemical Reaction?:'
                    '\na: color change'
                    '\nb: gas or bubbles'
                    '\nc: it becomes poisonous'
                    '\nd: creates heat/light'
                    '\ne: scent changes\nAnswer: ')

    while (answer3 != 'a' and answer3 != 'b' and answer3 != 'c' and answer3 != 'd' and answer3 != 'e'):
        answer3 = input('\nInvalid!\nEnter "a", "b", "c", "d", or "e" : ')

    if (answer == 't' and answer2 == 'b' and answer3 == 'c'):
        print('\nThat\'s three in a row!')
    elif (answer2 == 'b' and answer3 == 'c'):
        print('\nThat\'s two in a row!')    
    elif answer3 == 'c':
        print('Great job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')

    answer4 = input('\nProblem 4: An Ionic compound is the result of the bonding of:'
                    '\na: two metals'
                    '\nb: two nonmetals'
                    '\nc: a metal and nonmetal'
                    '\nd: two distinctly similar metals'
                    '\nAnswer: ')

    while (answer4 != 'a' and answer4 != 'b' and answer4 != 'c' and answer4 != 'd'):
        answer4 = input('\nInvalid!\nEnter "a", "b", "c",or "d": ')

    if (answer == 't' and answer2 == 'b' and answer3 == 'c' and answer4 == 'c'):
        print('\nThat\'s four in a row!')
    elif (answer3 == 'c' and answer4 == 'c'):
        print('\nThat\'s two in a row!')
    elif (answer2 == 'b' and answer3 == 'c' and answer4 == 'c'):
        print('\nThat\'s three in a row!')
    elif answer4 == 'c':
        print('Great job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')

    answer5 = input('\nProblem 5: What happens/is likely to happen if the atoms within an element '
                    'have full valence shells?:'
                    '\na: They are more likely to bond with other elements more easily'
                    '\nb: They are unlikely to bond with other elements'
                    '\nc: They will be labeled as "unstable"'
                    '\nd: They will explode'
                    '\nAnswer: ')

    while (answer5 != 'a' and answer5 != 'b' and answer5 != 'c' and answer5 != 'd'):
        answer5 = input('\nInvalid!\nEnter "a", "b", "c",or "d": ')

    if (answer == 't' and answer2 == 'b' and answer3 == 'c' and answer4 == 'c' and answer5 == 'b'):
        print('\nThat\'s all of them correct! Very impressive!')    
    elif (answer2 == 'b' and answer3 == 'c' and answer4 == 'c' and answer5 == 'b'):
        print('\nThat\'s four in a row!')
    elif (answer3 == 'c' and answer4 == 'c' and answer5 == 'b'):
        print('\nThat\'s three in a row!')
    elif (answer4 == 'c' and answer5 == 'b'):
        print('\nThat\'s two in a row!')
    elif answer5 == 'b':
        print('Great job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')
        
    # here we check if they want to try the quiz again
    play_again = input('Try the chemistry quiz again?\nEnter "yes" or "no" ')
    while (play_again != 'yes' and play_again != 'no'):
        play_again = input('\nInvalid!\nEnter "yes" or "no": ')
    # the answer is given by the player, then we change or keep the value stored in 'again', this will either satisfy
    # or fail to satisfy the requirements for continuing the game (keep in mind for a game to start it has to 
    #be: 'game' == True)
    if play_again == 'yes':
        again = True
    else:
        again = False
        
while topic == 'history' and again == True:
    answer = input('\nProblem 1:\nDeclaration of Independence is signed:'
                   '\na: 1776'
                   '\nb: 1773'
                   '\nc: 1769'
                   '\nd: 1770'
                   '\nAnswer: ')
    
    while (answer != 'a' and answer != 'b' and answer != 'c' and answer != 'd'):
        answer = input('\nInvalid!\nEnter "a", "b", "c" or "d": ')
    
    if answer == 'a':
        print('\nGreat job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')

    answer2 = input('\nProblem 2: When were the Crusades?: '
                '\na: 1300-1600'
                '\nb: 1100-1300'
                '\nc: 1000-1400'
                '\nd: 1200-1300'
                '\nAnswer: ')
    
    while (answer2 != 'a' and answer2 != 'b' and answer2 != 'c' and answer2 != 'd'):
        answer2 = input('\nInvalid!\nEnter "a", "b", "c" or "d": ')
    
    if (answer == 'a' and answer2 == 'b'):
        print('\nThat\'s two in a row!')
    elif answer2 == 'b':
        print('Great job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')
        
    answer3 = input('\nProblem 3: When did Columbus rediscover the Western Hemisphere?: '
                '\na: 1494'
                '\nb: 1490'
                '\nc: 1592'
                '\nd: 1492'
                '\nAnswer: ')
    
    while (answer3 != 'a' and answer3 != 'b' and answer3 != 'c' and answer3 != 'd'):
        answer3 = input('\nInvalid!\nEnter "a", "b", "c" or "d": ')
        
    if (answer == 'a' and answer2 == 'b' and answer3 == 'd'):
        print('\nThat\'s three in a row!')
    elif (answer2 == 'b' and answer3 == 'd'):
        print('\nThat\'s two in a row!')    
    elif answer3 == 'd':
        print('Great job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')
        
    answer4 = input('\nProblem 2: When did Columbus rediscover the Western Hemisphere?: '
            '\na: 1440-1455'
            '\nb: 1450-1465'
            '\nc: 1460-1465'
            '\nd: 1450-1455'
            '\nAnswer: ')
    
    while (answer4 != 'a' and answer4 != 'b' and answer4 != 'c' and answer4 != 'd'):
        answer4 = input('\nInvalid!\nEnter "a", "b", "c" or "d": ')
    
    if (answer == 'a' and answer2 == 'b' and answer3 == 'd' and answer4 == 'b'):
        print('\nThat\'s four in a row!')
    elif (answer2 == 'b' and answer3 == 'd' and answer4 == 'b'):
        print('\nThat\'s three in a row!')
    elif (answer3 == 'd' and answer4 == 'b'):
        print('\nThat\'s two in a row!')
    elif answer4 == 'b':
        print('Great job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')
    
    answer5 = input('\nProblem 2: Martin Luther protests against the Church?: '
            '\na: 1643'
            '\nb: 1215'
            '\nc: 1512'
            '\nd: 1517'
            '\nAnswer: ')
        
    while (answer5 != 'a' and answer5 != 'b' and answer5 != 'c' and answer5 != 'd'):
        answer5 = input('\nInvalid!\nEnter "a", "b", "c" or "d": ')
       
    if (answer == 'a' and answer2 == 'b' and answer3 == 'd' and answer4 == 'b' and answer5 == 'd'):
        print('\nThat\'s all of them correct! Very impressive!')
    elif (answer2 == 'b' and answer3 == 'd' and answer4 == 'b' and answer5 == 'd'):
        print('\nThat\'s four in a row!')
    elif (answer3 == 'd' and answer4 == 'b' and answer5 == 'd'):
        print('\nThat\'s three in a row!')
    elif (answer4 == 'b' and answer5 == 'd'):
        print('\nThat\'s two in a row!')
    elif answer5 == 'd':
        print('Great job! That is correct!')
    else:
        print('\nOh no! Unfortunately that is incorrect.')
        
    play_again = input('Try the history quiz again?\nEnter "yes" or "no" ')
    while (play_again != 'yes' and play_again != 'no'):
        play_again = input('\nInvalid!\nEnter "yes" or "no": ')
    if play_again == 'yes':
        again = True
    else:
        again = False
```
